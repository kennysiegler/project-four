```python
df = get_data('SPY',1999,1,1)
```

```python
# add features to model (SPY dataframe)
df['Range'] = ta.true_range(df['High'], df['Low'], df['Close'])
df['RSI 14'] = ta.rsi(df['Close'], 14)
df['VWAP'] = ta.vwap(df['High'], df['Low'], df['Close'], df['Volume'],1)
df['Momentum'] = ta.roc(df['Close'], 12)
df['ATR'] = ta.atr(df['High'], df['Low'], df['Close'])
df['Volume SMA 5'] = ta.sma(close=df['Volume'], length=5)
```

```python
df[-10:]
```

```python
# make new indicators 
# ratio of true range on that day to average true range
df['R ratio'] = df['Range']/df['ATR']

# did the day close above yesterday's close?
df['Higer Close'] = (df['Close'] > (df['Close'].shift(-1))).astype(int)

# did the day open above yesterday's close (did it gap up?)
df['Gap Up'] = (df['Open'] > (df['Close'].shift(-1))).astype(int)

# percent change on the day
df['Percent Change'] = ((df['Close']-df['Open'])/df['Open'])*100



# trend up
#df['Trend Up'] = (df['Close'] > df['Open'].shift(-1)).astype(int)

#df['Trend Up'] = 

df.drop(columns=['Adj Close', 'Open', 'High', 'Low', 'Close', 'Volume'])
```

```python
# create 'Tomorrow' column
# it's the 'Close' value from the previous day
# will be used to create a target
df['Tomorrow'] = df['Close'].shift(-1)
df['Target'] = (df['Tomorrow'] > df['Close']).astype(int)
df = df.loc['2000-01-01':]
df.drop('Adj Close', axis=1, inplace=True)
df.drop(df.tail(1).index, inplace=True)
```
